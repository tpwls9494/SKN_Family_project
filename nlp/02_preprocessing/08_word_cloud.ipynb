### 텍스트 전처리

```python
import pandas as pd
import numpy as np
from konlpy.tag import Okt
from collections import Counter
from wordcloud import WordCloud
import matplotlib.pyplot as plt
```

```python
spam_df = pd.read_csv('./sms_spam.csv')
spam_df.head()
```

```python
corpus = spam_df['CN'][0]
corpus
```

```python
okt = Okt()
nouns = okt.nouns(corpus)
nouns
```

```python
word_count = Counter(nouns)
word_count
```

```python
ko_stopwords = ['은', '는', '이', '가', '및', '더', '수']
word_count = {word:count for word, count in word_count.items() if word not in ko_stopwords}

word_count
```

### WordCloud 객체 생성

```python
wordcloud = WordCloud(
    width=800,
    height=800,
    font_path = '/Library/Fonts/Arial Unicode.ttf',
    background_color='white'
).generate_from_frequencies(word_count)
wordcloud
```

```python
plt.figure(figsize=(10, 10))
plt.imshow(wordcloud)
plt.axis('off')
plt.show()
```

```python
corpus = spam_df['CN']

okt = Okt()
nouns = []
for corpus_temp in corpus:
    nouns.extend(okt.nouns(corpus_temp))
```

```python
word_count = Counter(nouns)

ko_stopwords = ['은', '는', '이', '가', '및', '더', '수']
word_count = {word:count for word, count in word_count.items() if word not in ko_stopwords}
```

```python

wordcloud = WordCloud(
    width=800,
    height=800,
    font_path = '/Library/Fonts/Arial Unicode.ttf',
    background_color='white'
).generate_from_frequencies(word_count)
```

```python
plt.figure(figsize=(10, 10))
plt.imshow(wordcloud)
plt.axis('off')
plt.show()
```
